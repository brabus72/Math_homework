```python
def Descart_coor(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y

Descart_coor(3, (-2*np.pi)/3)
```

1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x(1 – y) = 1

```python
x = np.linspace(-10, 10, 1001)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5.5, 55)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equa(p):
    x,y = p
    return(y - x**2 + 1, y - ((np.exp(x) + x - 1)/x))
x1, y1 = fsolve(equa, (-2, 0))
x2, y2 = fsolve(equa, (1.5, 2.5))
x3, y3 = fsolve(equa, (3.5, 12.5))

print(x1, y1)
print(x2, y2)
print(x3, y3)
```

1) Напишите код, проверяющий любую из теорем сложения или умножения вероятности на примере подбрасывания монетки.

```python
# проверим сложение

def coin_toss(attempts):
    toss = np.random.rand(attempts)
    avers = sum(i > 0.5 for i in toss)
    revers = sum(i < 0.5 for i in toss)
    summ = len(toss)
    print(f' Вероятность выпадения "орла" монеты = {avers/100} \n',
          f'Вероятность выпадения "решки" монеты = {revers/100} \n',
          f'Сумма вероятностей подбрасывания монеты = {summ/100}')
    
coin_toss(100)
```

Дополните код расчетом коэффициента корреляции x и y по формуле

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

R = sum((x - x.mean())*(y - y.mean())) / (np.sqrt(sum(np.square(x - x.mean()))*sum(np.square(y - y.mean()))))

plt.scatter(x, y)
plt.plot([0, 1], [b, a + b], 'r')
plt.grid(alpha=0.5, linestyle='--')
plt.show()
print(R)
```

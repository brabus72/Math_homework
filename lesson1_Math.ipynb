Задание 3

Напишите код на Python, реализующий построение графиков: 1. окружности, 2. эллипса, 3. гиперболы.

```python
R = 5
x = np.linspace(-R, R, 1000)
x1 = []
y1 = []
y2 = []

for i in x:
    x1.append(i)
    y1.append(np.sqrt(R**2 - i**2))
    y2.append(- np.sqrt(R**2 - i**2))

plt.figure(figsize=plt.figaspect(1.0))
plt.plot(x1,y1, color='red')
plt.plot(x1,y2, color='black')

plt.xlabel("x")
plt.ylabel("y");
```

```python
x = np.linspace(-15, 15, 10000)
a = 5
b = 9

y = np.sqrt((1 - x**2/a**2)/b**2)
y1 = -np.sqrt((1 - x**2/a**2)/b**2)

plt.plot(x,y, color='green')
plt.plot(x,y1, color='green')
plt.xlabel("x")
plt.ylabel("y");
```

Задание 5

1) Нарисуйте трехмерный график двух параллельных плоскостей. 
2) Нарисуйте трехмерный график двух любых поверхностей второго порядка. 

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(X, Y)
Z = 3*X + 5*Y
Z1 = 3*X + 5*Y + 100
ax.plot_surface(X,Y,Z, color='green')
ax.plot_wireframe(X,Y,Z1, color='yellow');
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2 + Y**2)
Z1 = - np.sqrt(X**2 + Y**2)
ax.plot_wireframe(X,Y,Z, color='red')
ax.plot_surface(X,Y,Z1, color='red')
ax.scatter(0, 0, 0, 'z', 20, 'black')
show();
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(X, Y)
a = 3
b = 3
c = 3
Z = np.sqrt((X**2/a**2 + Y**2/b**2 + 1)/c**2)
Z1 = - np.sqrt((X**2/a**2 + Y**2/b**2 + 1)/c**2)
ax.plot_wireframe(X,Y,Z, color='blue')
ax.plot_surface(X,Y,Z1, color='blue')
ax.scatter(0, 0, 0, 'z', 20, 'black');
show();
```
